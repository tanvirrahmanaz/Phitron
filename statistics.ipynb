```python
df = pd.DataFrame({
    "score": [10, 12, 8, 15, 7, 11, 9, 14]
})

print(df)
```

```python
mean = df['score'].sum()/ len(df)
mean
```

```python
mean = df['score'].mean()
df['dev'] = df['score'] - mean
df['sq_dev'] = df['dev'] **2

population_varience = df['sq_dev'].sum() / len(df)

print(df)
population_varience
```

```python
sample_variance = df['sq_dev'].sum()/(len(df) - 1)
sample_variance
```

```python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
plt.scatter(df.index, df['score'], s=100)
plt.axhline(mean, color='red', linestyle='--', label = f'Mean = {mean: .2f}')


for i , val in enumerate(df['score']):
    plt.plot([i,i],[mean,val],color='gray', linestyle='--')

plt.title("deviation from mean(variance visualization)")
plt.xlabel('index')
plt.ylabel('score')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))

bars = plt.bar(df.index,df['sq_dev'], color='orange')
plt.title("Squared Deviations (Contribution to Variance)")
plt.xlabel("Index")
plt.ylabel("Squared Deviation")
plt.grid()


for i, val in enumerate(df['sq_dev']):
    plt.text(i, val + 0.2,      # position (x=i, y=value এর একটু উপরে)
             f"{val:.2f}",      # value format
             ha='center',       # horizontal alignment
             fontsize=10)
plt.show()
```
