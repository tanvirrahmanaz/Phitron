```python
import pandas as pd

# Example: Creating a larger and more detailed sales DataFrame
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales_A': [220, 270, 300, 350, 400, 420, 450, 470, 490, 510, 530, 550],
    'Sales_B': [180, 230, 280, 310, 360, 390, 420, 440, 460, 490, 510, 530],
    'Sales_C': [150, 200, 250, 290, 330, 360, 390, 410, 430, 460, 480, 500]
}

# Create DataFrame
df = pd.DataFrame(data)

# Add Total Sales & Profit columns
df['Total_Sales'] = df['Sales_A'] + df['Sales_B'] + df['Sales_C']
df['Profit'] = (df['Total_Sales'] * 0.25).round(2)  # Assume 25% profit margin

```

```python
df
```

```python
df.plot(x='Month', y = ['Sales_A','Sales_B'], kind='line', marker='^', linewidth=2)

plt.title("Monthly sales comparison")

plt.xlabel("Month")
plt.ylabel('Sales in units')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.plot([1,2,3], [4,5,6])
plt.title("Left Plot")

plt.subplot(1,2,2)
plt.plot([1,2,3], [6,5,4])
plt.title("Right Plot")

plt.tight_layout()  # <-- adjusts spacing automatically
plt.show()
```

```python
plt.style.use('seaborn-v0_8-darkgrid')

ax = df.plot(x='Month', y=['Sales_A', 'Sales_B'], marker='o', linewidth=3)
ax.set_title('Sales Performance - Styled Plot', fontsize=14, fontweight='bold')
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Sales (Units)', fontsize=12)
ax.legend(['Sales_A', 'Sales_B'], title='Product')
plt.show()
```
