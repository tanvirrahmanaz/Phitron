```python
def square_addition(a,b,c):
    print(f"a= {a}, b= {b}, c= {c}")
    a = a**2
    b = b**2
    c= c**2

    return a+b+c;

result = square_addition(b=2, c=4, a=3)
print(f"Final resutl is {result}")
```

```python
def square_addition(*args):
    summation =0 
    for i in args:
        print(f"{i}**2 = {i**2}")
        i = i**2
        summation += i
    return summation
    

result = square_addition(2,3,4)
print(f"Final resutl is {result}")

#result = square_addition(b=2, c=4, a=3) is not acceptable
```
